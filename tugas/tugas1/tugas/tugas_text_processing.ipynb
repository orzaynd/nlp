```python
#fungsi untuk membersihkan data full_text
def clean_full_text(df, column_name):

    df['full_text_clean'] = df[column_name].str.lower()

    df['full_text_clean'] = df['full_text_clean'].str.strip()

    df['full_text_clean'] = df['full_text_clean'].str.replace('[^\w\s]', '', regex=True)

    df['full_text_clean'] = df['full_text_clean'].str.replace(r'(.)\1{2,}', r'\1', regex=True)

    return df
```

```python
#import library yang dibutuhkan
import re
import json
import requests
```

```python
# crawl data slang
slang_url = "https://raw.githubusercontent.com/louisowen6/NLP_bahasa_resources/master/combined_slang_words.txt"
response = requests.get(slang_url)
slang_dict = json.loads(response.text)
print(slang_dict)
```

```python
#import library sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
#using sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
def clean_full_text(text):

    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    words = text.split()

    # normalisasi slang
    words = [slang_dict[word] if word in slang_dict else word for word in words]


    text = " ".join(words)

    # stemming
    text = stemmer.stem(text)

    return text
```

```python
# ==============================
# 8. Jalankan Pipeline
# ==============================
df['clean_full_text'] = df['full_text'].apply(clean_full_text)

print(df[['full_text', 'clean_full_text']])
```

```python
import spacy
nlp = spacy.blank("id")  # tokenizer saja

def spacy_tokenize(text):
    doc = nlp(text)
    return [token.text for token in doc]

df['tokens'] = df['clean_full_text'].apply(spacy_tokenize)
```
