Buatkan sebuah function yang mempunyai parameter `name`, dan `hobi`. Function tersebut akan mengembalikan sebuah kalimat
**nama saya `name`, saya suka `hobi`.**

```python
class may_self:
    def __init__(self, name, hobby):
        self.name = name
        self.hobby = hobby
    def introduction(self):
        return(f"Nama saya {self.name}, saya suka {self.hobby}")
```
